```python
# Simplest One!!!
import numpy as np
from skimage.measure import perimeter

def estimated_circularity(mask):
    area = np.sum(mask)
    perim = perimeter(mask)
    circularity = (4 * np.pi * area) / (perim ** 2)
    return circularity

# Assuming sam_result is your list of dictionaries with actual data
# sam_result = [...]  # Uncomment and replace with your actual data

# List to store indices of segments with the specified area and circularity
selected_indices = []

# Iterate over the segments
for idx, segment in enumerate(sam_result):
    mask = segment['segmentation']
    area = segment['area']
    area_scaled = area / 1000  # Scale the area as per your requirement

    # Check if the area is between 40 and 50 and circularity is >= 0.5
    if 40 <= area_scaled <= 50 and estimated_circularity(mask) >= 0.5:
        selected_indices.append(idx)

# Output the selected indices
print(selected_indices)


```

```python
# Remove overlapping
def calculate_overlap(mask1, mask2):
    intersection = np.logical_and(mask1, mask2)
    union = np.logical_or(mask1, mask2)
    if np.sum(union) == 0:
        return 0
    return np.sum(intersection) / np.sum(union)

len_selected_indices = len(selected_indices)

overlap_matrix = np.zeros((len_selected_indices, len_selected_indices))

copied_selected_indices = selected_indices.copy()

for i in range(len_selected_indices-1):
    for j in range(i+1, len_selected_indices):
        mask1 = sam_result[selected_indices[i]]['segmentation']
        mask2 = sam_result[selected_indices[j]]['segmentation']
        
        overlap_percent = 100*calculate_overlap(mask1, mask2)
        if overlap_percent > 50:
            area_mask1 = sam_result[selected_indices[i]]['area']
            area_mask2 = sam_result[selected_indices[j]]['area']
            
            
            if area_mask1 > area_mask2:
                removal_index = j
            else:
                removal_index = i
            
            copied_selected_indices.pop(removal_index)
            
print(copied_selected_indices)

sam_selected_indices = [sam_result[i] for i in selected_indices]

# Example usage
_, axes = plt.subplots(1, 2, figsize=(16, 16))
# Replace 'image_rgb' with your actual RGB image
axes[0].imshow(image_rgb)
show_output2(sam_selected_indices, axes[1])

# # Visualize the selected segments
plt.figure(figsize=(10, 10))
for i, idx in enumerate(selected_indices):
    mask = sam_result[idx]['segmentation']

    plt.subplot(3, 3, i + 1)  # Adjust the layout based on the number of segments
    plt.imshow(mask, cmap='gray')
    plt.title(f'Segment {idx}')
    if i == 8:  # Adjust based on how many segments you want to display
        break

plt.tight_layout()
plt.show()
len(selected_indices)
```

```python
# Function that inputs the output and plots image and mask with area
def show_output2(result_dict, axes=None):
    if axes:
        ax = axes
    else:
        ax = plt.gca()
        ax.set_autoscale_on(False)

    sorted_result = sorted(result_dict, key=(lambda x: x['area']), reverse=True)

    # Plot for each segment area
    for val in sorted_result:
        mask = val['segmentation']
        area = val['area']/1000

        # Generate a random color for the mask
        color_mask = np.random.random((1, 3)).tolist()[0]
        img = np.ones((mask.shape[0], mask.shape[1], 3))
        for i in range(3):
            img[:, :, i] = color_mask[i]

        ax.imshow(np.dstack((img, mask * 0.5)))

        # Find the center of the mask to place the text
        y, x = np.where(mask)
        center_x, center_y = np.mean(x), np.mean(y)

        # Add the area text on the mask
        ax.text(center_x, center_y, f'{area:.2f}', color='white', ha='center', va='center', fontsize=8)

# Example usage
_, axes = plt.subplots(1, 2, figsize=(16, 16))
# Replace 'image_rgb' with your actual RGB image
axes[0].imshow(image_rgb)
show_output2(sam_result, axes[1])
```

```python
# Experiment Hough transform
import cv2
import numpy as np
import matplotlib.pyplot as plt

def find_circles(segmentation, min_dist, min_radius, max_radius):
    # Apply Hough Circle Transform
    circles = cv2.HoughCircles(segmentation, cv2.HOUGH_GRADIENT, dp=1, minDist=min_dist,
                               param1=50, param2=30, minRadius=min_radius, maxRadius=max_radius)
    return circles

circular_objs_info = []

# Parameters for Hough Circle Transform (adjust these based on your data)
min_dist = 100  # Minimum distance between detected centers
min_radius = 40  # Minimum radius to be detected
max_radius = 50 # Maximum radius to be detected

# Visualize segments identified as circles
plt.figure(figsize=(10, 10))
for i, segment in enumerate(sam_result):
    mask = segment['segmentation']
    mask = (mask * 255).astype('uint8')  # Convert mask to uint8 type
    circles = find_circles(mask, min_dist, min_radius, max_radius)
    print(circles)
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for circle in circles[0,:]:
            center = (circle[0], circle[1])
            radius = circle[2]
            cv2.circle(mask, center, radius, (255, 0, 0), 2)
            print('here')
#     circular_objs_info.append({"segment_index": i, "radius": radius, "area": area})
    plt.subplot(3, 3, i+1)
    plt.imshow(mask, cmap='gray')
    plt.title(f'Segment {i+1}')
    if i == 8:  # Adjust based on how many segments you want to display
        break

plt.tight_layout()
plt.show()

print(circular_objs_info)
```

```python
import numpy as np
from skimage.measure import perimeter

def estimated_circularity(mask):
    area = np.sum(mask)
    perim = perimeter(mask)
    circularity = (4 * np.pi * area) / (perim ** 2)
    return circularity

def group_segments_by_area(segments, area_tolerance=0.05):
    groups = {}
    for idx, segment in enumerate(segments):
        area = segment['area']
        found_group = False
        for group_area in groups.keys():
            if abs(area - group_area) / group_area <= area_tolerance:
                groups[group_area].append(idx)
                found_group = True
                break
        if not found_group:
            groups[area] = [idx]
    return groups

```
